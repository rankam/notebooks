```python
x = np.arange(-10,10.01,.01)
y = 1 - x**2 + np.random.normal(0,5,len(x))
```

<h3>Visualizing Nonlinear Data</h3>

```python
plt.scatter(x,y)
```

```python
x_squared = x**2
```

```python
df = pd.DataFrame.from_dict({'x_squared':x_squared, 'y':y, 'x':x})
ggplot(aes(x='x_squared', y='y'),data=df) + geom_point() + stat_smooth(method='lm',color='red')
```

Plotting y against this new x_squared input gives us a fit that looks exactly like a straight line. Essentially, we've transformed our riignal nonlinear problem into a new problem in which the relationship between the two inputs really satasfies the linearity assumption of linear regression. This idea of replacing a complicated nonlinear problem iwth a simpler linear one using a transformation of the inputs comes up anagain and again in machine learning.

Let's see how much the squaring trtansformation imporoves our prediction quality by looking at our R<sup id="superscript">2</sup> values for linear regression using x and x_squared.

```python
print'y ~ x: ',smf.ols('y ~ x', data=df).fit().rsquared
print 'y ~ x^2: ',smf.ols('y ~ x_squared', data=df).fit().rsquared
```

With the transformation, we've gone from accounting for 0% of the variance to accounting for 97%. 

<h3>Polynomial Regression</h3>
Let's start with a simple example and show what happens as we give our model more expressive power to mimic the structure in our data. 

```python
np.random.seed(1)

x = np.arange(0,1,.01)
y = np.sin(2 * np.pi * x) + np.random.normal(0,0.1,len(x)) 
df = pd.DataFrame.from_dict({'y':y, 'x':x})
ggplot(aes('x','y'), data=df) + geom_point()
```

Looking at the data, you can tell that simple linear regression will not work, but let's see how it performs.

```python
smf.ols('y ~ x', data=df).fit().summary2()
```

We're able to explain 60% of the variance in the data even though we know that this linear regression model is a bad model of wave data. We also know that a good model should be able to explain more than 90% of the variance in this data set. However, let's see how this linear model was able to produce such a fit to the data. 

```python
ggplot(aes('x','y'), data=df) + geom_point() + stat_smooth(method='lm', se=False)
```

We can see that the linear model was able to capture the wave's downward structure - however, if the wave continued through another period, the linear model would capture nearly 0% of the variance. 

What happens when we add new features to our data set?

```python
df['x2'] = df['x']**2
df['x3'] = df['x']**3
smf.ols('y ~ x + x2 + x3', data=df).fit().summary2()
```

The additional features changed our R^2 from 60% to 97%. We could continue to add features such as x^4, x^5 ...x^n and one may think that the our model would continue to improve. The reality is that by doing this, we will add new features with larger and larger powers of x. The new features will be so correlated with the old features that the linear regression algorithm will break down and not be able to find coefficients for all of the features separately. When problem arises, it is known as a singularity. 

The solution to this problem can be achieved by adding more complicated variants of x that work like successive powers of x, but aren't correlated with each other in the way that x and x^2 are. These variants on the powers of x are called orthogonal polynomials.

```python
new_df = patsy.dmatrix("C(x, Poly)", data=df, return_type='dataframe').iloc[:,:15]
new_df['y'] = y
new_df.drop('Intercept',axis=1, inplace=True)
new_df.head()
```

```python
# s = ' + '.join(new_df.drop('y',axis=1).columns)
smf.ols('y ~ x', data=new_df).fit().summary()
```

```python
from sklearn.feature_extraction.text import CountVectorizer 

def fn_tdm_df(docs, xColNames = None, **kwargs):
    ''' create a term document matrix as pandas DataFrame
    with **kwargs you can pass arguments of CountVectorizer
    if xColNames is given the dataframe gets columns Names'''

    #initialize the  vectorizer
    vectorizer = CountVectorizer(**kwargs)
    x1 = vectorizer.fit_transform(docs)
    #create dataFrame
    df = pd.DataFrame(x1.toarray().transpose(), index = vectorizer.get_feature_names())
    if xColNames is not None:
        df.columns = xColNames

    return df
```
