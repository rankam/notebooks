<h2>Using Patsy to Determine Which Team Michael Jordan Beat the Most</h2>
<p>
In this simple exercise, I use the Patsy library to quickly convert categorical data (Michael Jordan's opponent & the game result) into binary features so that I can aggregate the number of wins Michael Jordan had against each team.</p>

```python
%pylab inline
figsize(12,8)
import numpy as np
import matplotlib.pyplot as plt
import patsy
plt.style.use('fivethirtyeight')
```

```python
mj = pd.read_csv('/users/aaronrank/dropbox/mj_game_stats.csv')
mj.columns = [col.lower() for col in mj.columns]
mj.head()
```

```python
mj = mj.rename(columns={'3pa':'three_pa', '3p':'three_p'}) # Patsy was parsing 3pa as 3 pa
s = " + ".join([col for col in mj.columns if  col != 'date']) + "+ opponent:result -1" # Patsy would treat 'date' as categorical
mj_patsy = patsy.dmatrix(s, mj, return_type='dataframe')
mj_patsy.head()
```

Using Patsy's ability to create higher order terms, I've created additional binary features that return 1 if Michael Jordan's team beat their opponent and 0 if Michael Jordan's team lost. I can now take the sum of each of my newly created features to return the number of wins Michael Jordan had against each team. 

```python
mj_wins = pd.DataFrame(mj_patsy[[col for col in mj_patsy if str(col).startswith('opponent') and str(col).endswith('W]')]].sum().order()).T
mj_wins.columns = [team[11:-13:] for team in mj_wins.columns]
mj_wins
```

Using Python's built in startswith() and endswith() functions, I can rename each column from opponent[Team Name] to Team Name making any forthcoming analysis or visualization more easy to interpret.

```python
fig = pyplot.figure()
figsize(12,8)
plt.bar(np.arange(len(mj_wins.columns)), mj_wins.values[0],color='teal')
plt.xlim(0,31)
plt.ylim(0,50)
plt.ylabel('Wins')
plt.title('Career Wins Against Teams')
plt.xticks(np.arange(len(mj_wins.columns)) + .4, mj_wins.columns)
plt.xticks(rotation=90)
plt.xlabel('Teams',labelpad=290)
plt.subplots_adjust(bottom=0.40)
```
