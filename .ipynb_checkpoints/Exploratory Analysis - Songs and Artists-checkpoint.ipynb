### Workflow for Exploratory Analysis - Based off of Harvard's CS109: Data Science
1. **Build** a DataFrame from the data (ideally, put *all* data in this object)
2. **Clean** the DataFrame. It should have the following properties:
   * Each row describes a single object
   * Each column describes a property of that object
   * Columns are numeric whenever appropriate
   * Columns contain atomic properties that cannot be further decomposed
3. Explore **global properties**. Use histograms, scatter plots, and aggregation functions to summarize the data.
4. Explore **group properties**. Use groupby and small multiples to compare subsets of the data.

Doing this will transform my data into a format which is easier to work with, gives me a basic overview of 
the data's properties, and likely generates several questions for me to followup in subsequent analysis.

```python
%pylab inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import futil

#tell pandas to display wide tables as pretty HTML tables
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

## Build a DataFrame

```python
artist_term = pd.read_csv('artist_term.csv')
artist_term.head()
```

```python
songs = pd.read_csv('songs.csv')
songs.head()
```

```python
data = songs.merge(artist_term, on='artist_id',how='left')
data.head()
```

## 2. Clean the Data

```python
data = data.dropna(how='any')
data[data['year'] == 0] = np.nan
```

## 3. Explore global properties

Next, we get a handle on some basic, global summaries of the DataFrame.

```python
data.describe()
```

<h3>Make some basic plots</h3>

```python
data['duration'].plot(kind='kde', color='r')
xlabel('Duration')
remove_border()
```

```python
data['artist_familiarity'].plot(kind='kde')
xlabel('Artist Familiarity')
```

```python
data['artist_hotttnesss'].plot(kind='kde')
xlabel('Artist Hotness')
remove_border()
```

```python
figsize(8,4.5)
plt.scatter(data['year'], data.artist_familiarity, lw=0, alpha=.08, color='g')
ylabel('Artist Hotness')
xlabel('Year')
remove_border()
```

### Run aggregation functions like `sum` over several rows or columns

```python
genres = dict()
ct = 1
for genre in data['term']:
    if genre not in genres:
        genres[genre] = ct
    else:
        genres[genre] += 1
pd.DataFrame({'Genre Count': genres}).sort('Genre Count',ascending=False).head(10)
```

```python
figsize(12,6.5)
plt.style.use('fivethirtyeight')
pd.DataFrame({'Genre Count': genres}).sort('Genre Count',ascending=False).head(10).plot(kind='bar')
plt.subplots_adjust(bottom=0.30)
remove_border()
title('Top 10 Music Genres')
```

*How many genres does a song have, on average?*

```python
genre_count = data.groupby(['title'])['term'].count('term').reset_index()
print "Average song has %0.2f genres" % genre_count.mean()
genre_count.describe()
```

<h3>Explore Group Properties</h3>

Let's split music by decade

```python
decade =  (data.year // 10) * 10
tyd = data[['term', 'year']]
tyd['decade'] = decade
tyd.head()
```

```python
data['decade'] = decade
decade_mean = data.groupby(decade).artist_hotttnesss.mean()
decade_mean.name = 'Decade Hotness'
print decade_mean

plt.plot(decade_mean.index, decade_mean.values, 'o-',
        color='r', lw=3, label='Decade Average')
plt.scatter(data.year, data.artist_hotttnesss, alpha=.04, lw=0, color='k')
plt.xlabel("Year")
plt.ylabel("Hotness")
plt.legend(frameon=False)
remove_border()
plt.subplots_adjust(bottom=0.10)
```

```python
figsize(12,6.5)
rap_mean = data[data['term']=='rap'].groupby(decade).artist_hotttnesss.mean()
rock_mean = data[data['term']=='rock'].groupby(decade).artist_hotttnesss.mean()
pop_mean = data[data['term']=='pop'].groupby(decade).artist_hotttnesss.mean()
country_mean = data[data['term']=='country'].groupby(decade).artist_hotttnesss.mean()

decade_mean.name = 'Decade Hotness'
print decade_mean

plt.plot(rap_mean.index, rap_mean.values, '-o',
        color='r', lw=3, label='Rap Average')
plt.plot(rock_mean.index, rock_mean.values, '-o',
        color='k', lw=3, label='Rock Average')
plt.plot(pop_mean.index, pop_mean.values, '-o',
        color='g', lw=3, label='Pop Average')
plt.plot(country_mean.index, country_mean.values, '-o',
        color='y', lw=3, label='Country Average')
plt.scatter(data.year, data.artist_hotttnesss, alpha=.01, lw=0, color='orange')
plt.xlabel("Year")
plt.ylabel("Hotness")
plt.legend(frameon=False)
remove_border()
plt.subplots_adjust(bottom=0.10)
```
