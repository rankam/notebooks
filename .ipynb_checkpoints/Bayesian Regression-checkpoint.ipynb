<h2>Comparing OLS and Bayesian Regression</h2>
<p>Based off of Chapter 5 of Cam Davidson-Pilon's <a href="http://nbviewer.ipython.org/github/CamDavidsonPilon/Probabilistic-Programming-and-Bayesian-Methods-for-Hackers/blob/master/Chapter5_LossFunctions/LossFunctions.ipynb">Probabilistic Programming and Bayesian Methods for Hackers</a>, I compare Ordinary Least Squares and Bayesian regression in predicting the number of points Michael Jordan's team will score based on how many points Michael Jordan scores.</p>

```python
%pylab inline
import numpy as np
import pandas as pd
import pymc as pm
plt.style.use('bmh')
figsize(12,6.5)
```

```python
mj = pd.read_csv('/users/aaronrank/dropbox/mj_game_stats.csv')
mj.columns = [col.lower() for col in mj.columns]
mj.head()
```

```python
mj['pts'].plot(kind='kde',label='Points - Michael Jordan')
mj['mj_score'].plot(kind='kde', label='Points - MJ\'s Team')
legend()
```

```python
N = len(mj['pts'])    
Y = mj['mj_score']
X = mj['pts']

ls_coef_ = np.cov(X,Y)[0,1] / np.var(X)
ls_intercept = Y.mean() - ls_coef_ * X.mean()

plt.scatter(X, Y, c='k')
plt.xlabel('MJ Points Scored')
plt.ylabel("MJ's Team Score")
plt.plot(X, ls_coef_ * X + ls_intercept, label='Least-sqares line')
plt.xlim(X.min(), X.max())
plt.ylim(Y.min(), Y.max())
plt.legend(loc='upper left')
plt.subplots_adjust(bottom=0.15)
```

```python
import pymc as pm
from pymc.Matplot import plot as mcplot

std = pm.Uniform('std', 0, 100, trace=False)

@pm.deterministic
def prec(U=std):
    return 1./ (U) ** 2

beta = pm.Normal('beta', 0,0.0001)
alpha = pm.Normal('alpha', 0,0.0001)

@pm.deterministic
def mean(X=X, alpha=alpha, beta=beta):
    return alpha + beta * X

obs = pm.Normal('obs', mean, prec, value=Y, observed=True)
mcmc = pm.MCMC([obs, beta, alpha, std, prec])

mcmc.sample(100000, 80000)
mcplot(mcmc)
```

```python
from scipy.optimize import fmin


def points_scored(points, pred, coef=500):
    """vectorized for numpy"""
    sol = np.zeros_like(points)
    ix = points * pred < 0
    sol[ix] = coef * pred ** 2 - np.sign(points[ix]) * pred + abs(points[ix])
    sol[~ix] = abs(points[~ix] - pred)
    return sol

tau_samples = mcmc.trace("prec")[:]
alpha_samples = mcmc.trace("alpha")[:]
beta_samples = mcmc.trace("beta")[:]

N = tau_samples.shape[0]

noise = 1. / np.sqrt(tau_samples) * np.random.randn(N)

possible_outcomes = lambda signal: alpha_samples + beta_samples * signal \
    + noise


opt_predictions = np.zeros(50)
scoring_signals = np.linspace(X.min(), X.max(), 50)
for i, _signal in enumerate(scoring_signals):
    _possible_outcomes = possible_outcomes(_signal)
    tomin = lambda pred: points_scored(_possible_outcomes, pred).mean()
    opt_predictions[i] = fmin(tomin, 0, disp=False)


plt.xlabel("scoring signal")
plt.ylabel("prediction")
plt.title("Least-squares prediction vs. Bayes action prediction")
plt.plot(X, ls_coef_ * X + ls_intercept, label="Least-squares prediction")
plt.xlim(X.min(), X.max())
plt.plot(scoring_signals, opt_predictions, label="Bayes action prediction")
plt.legend(loc="upper left");
```
