<h2>Inferring Behavior Based on Scoring</h2>
<p>Based off of Chapter 1 of Cam Davidson-Pilon's <a href="http://nbviewer.ipython.org/github/CamDavidsonPilon/Probabilistic-Programming-and-Bayesian-Methods-for-Hackers/blob/master/Chapter1_Introduction/Chapter1_Introduction.ipynb">Probabilistic Programming and Bayesian Methods for Hackers</a>, I explore if Michael Jordan's scoring changed throughout his career in the NBA.</p>

```python
%pylab inline
figsize(12,8)
import numpy as np
import scipy.stats as stats
import pandas as pd
import psycopg2
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
from matplotlib import dates
plt.style.use('fivethirtyeight')
```

```python
mj = pd.read_csv('/users/aaronrank/dropbox/mj_game_stats.csv')
mj.columns = [col.lower() for col in mj.columns]
```

```python
figsize(13.5,5.5)
n_count_data = len(mj['pts'])
plt.bar(np.arange(n_count_data), mj['pts'],color='#348ABD')
plt.xlabel("Time (Game)")
plt.ylabel("Points Scored Per Game")
plt.title("Did Michael Jordan's Scoring Habits Change Over Time?")
plt.subplots_adjust(bottom=0.15)
```

```python
import pymc as pm
n_count_data = len(mj['pts'])
alpha = 1. / mj['pts'].mean()
lambda_1 = pm.Exponential('lambda_2', alpha)
lambda_2 = pm.Exponential('lambda_1', alpha)
tau = pm.DiscreteUniform('tau', lower=0, upper=n_count_data)
```

```python
print 'Random output:', tau.random(), tau.random(), tau.random()
```

```python
@pm.deterministic
def lambda_(tau=tau, lambda_1=lambda_1, lambda_2=lambda_2):
    out = np.zeros(n_count_data)
    out[tau:] = lambda_1
    out[:tau] = lambda_2
    return out
```

```python
observation = pm.Poisson('obs', lambda_, value=mj['pts'], observed=True)
model = pm.Model([observation, lambda_1, lambda_2, tau])
```

```python
mcmc = pm.MCMC(model)
mcmc.sample(40000, 10000, 1)
```

```python
lambda_1_samples = mcmc.trace('lambda_1')[:]
lambda_2_samples = mcmc.trace('lambda_2')[:]
tau_samples = mcmc.trace('tau')[:]
```

```python
figsize(12.5, 10)
# # histogram of the samples:
plt.style.use('bmh')

ax = plt.subplot(311)

plt.hist(lambda_1_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\lambda_1$", color="#A60628", normed=True)
plt.legend(loc="upper left")
plt.title(r"""Posterior distributions of the variables
    $\lambda_1,\;\lambda_2,\;\tau$""")
plt.ylim([0, 2.9])
plt.xlim([25, 40])
plt.xlabel("$\lambda_1$ value")
plt.subplots_adjust(bottom=0.20)

ax = plt.subplot(312)
plt.hist(lambda_2_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\lambda_2$", color="#7A68A6", normed=True)
plt.legend(loc="upper left")
plt.xlim([15, 25])
plt.ylim([0, 1.7])
plt.xlabel("$\lambda_2$ value")

plt.subplot(313)
w = 1.0 / tau_samples.shape[0] * np.ones_like(tau_samples)
plt.hist(tau_samples,alpha=1,
         label=r"posterior of $\tau$",
         color="#467821", weights=w, rwidth=2.)
plt.xticks(np.arange(n_count_data))



plt.legend(loc="upper left")
plt.xlim([830,850])
plt.xlabel(r"$\tau$ (# games)")
plt.ylabel("probability")
plt.subplots_adjust(bottom=0.10)
```

```python
figsize(12.5, 5)
# tau_samples, lambda_1_samples, lambda_2_samples contain
# N samples from the corresponding posterior distribution
N = tau_samples.shape[0]
expected_ppg = np.zeros(n_count_data)
for game in range(0, n_count_data):
    # ix is a bool index of all tau samples corresponding to
    # the switchpoint occurring prior to value of 'day'
    ix = game < tau_samples
    # Each posterior sample corresponds to a value for tau.
    # for each day, that value of tau indicates whether we're "before"
    # (in the lambda1 "regime") or
    #  "after" (in the lambda2 "regime") the switchpoint.
    # by taking the posterior sample of lambda1/2 accordingly, we can average
    # over all samples to get an expected value for lambda on that day.
    # As explained, the "message count" random variable is Poisson distributed,
    # and therefore lambda (the poisson parameter) is the expected value of
    # "message count".
    expected_ppg[game] = (lambda_1_samples[ix].sum()
                                   + lambda_2_samples[~ix].sum()) / N


plt.plot(range(n_count_data), expected_ppg, lw=4, color="#E24A33",
         label="Expected Points Per Game")
plt.xlim(0, n_count_data)
plt.xlabel("Game")
plt.ylabel("Expected # Points")
plt.title("Expected Points Per Game")
plt.ylim(0, 60)
plt.bar(np.arange(len(mj['pts'])), mj['pts'], color="#348ABD", alpha=0.65,
        label="Actual Points Per Game")
plt.legend(loc="upper left")
plt.subplots_adjust(bottom=0.15)
```

<h3>Interpretation</h3>
The analysis shows strong support for believing that Michael Jordan's scoring behavior did change (λ1 would have been close in value to λ2 had this not been true), and that the change was sudden rather than gradual (as demonstrated by τ's strongly peaked posterior distribution). Michael Jordan's 840th regular season game was 10 games into his first season with the Washington Wizards where he had recently come out of retirement for the second time after a nearly 3 year hiatus. We can speculate that his age (He was nearing 39 years old) and his recent break from basketball may have been factors that caused his significant decrease in scoring per game.
