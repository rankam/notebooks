```julia
%pylab inline
figsize(8,4.5)
import pandas as pd
import numpy as np
import scipy.stats as stats
plt.style.use('bmh')
import statsmodels.formula.api as smf
```

```julia
data = pd.read_csv('top_1000_sites.tsv',sep='\t')
data.head()
```

```julia
plt.scatter(data['PageViews'], data['UniqueVisitors'])
plt.xlabel('Page Views')
plt.ylabel('Unique Visitors')
```

```julia
data['PageViews'].plot(kind='kde')
plt.title('KDE Page Views')
```

```julia
np.log(data['PageViews']).plot(kind='kde')
plt.title('KDE Log(PageViews)')
xlim(15,27)
```

```julia
scatter(np.log(data['UniqueVisitors']),np.log(data['PageViews']))
plt.xlabel('Log Page Views')
plt.ylabel('Log Unique Visitors')
ylim(15,28)
```

```julia
results = smf.ols('np.log(PageViews) ~ np.log(UniqueVisitors)', data=data).fit()
results.summary2()
```

```julia
data['InEnglish'] = data['InEnglish'].replace(np.nan,'Not Available') # necessary so that NAs are not removed
results = smf.ols(formula='np.log(PageViews) ~ HasAdvertising + np.log(UniqueVisitors) + C(InEnglish)' , data=data).fit()
results.summary2()
```

```julia
results = smf.ols(formula='np.log(PageViews) ~ HasAdvertising' , data=data).fit()
print 'HasAdvertising: ',results.rsquared
results = smf.ols(formula='np.log(PageViews) ~ np.log(UniqueVisitors)' , data=data).fit()
print 'log(UniqueVisitors): ',results.rsquared
results = smf.ols(formula='np.log(PageViews) ~ C(InEnglish)' , data=data).fit()
print 'InEnglish: ',results.rsquared
```
